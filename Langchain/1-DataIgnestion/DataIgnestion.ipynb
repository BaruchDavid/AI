RAG:

- Retrieval → Abruf von externen Informationen (z. B. aus einer Datenbank, Dokumenten oder dem Web).
- Augmented → Diese Informationen werden in den Generierungsprozess „eingespeist“ und erweitern das Modellwissen.
- Generation → Das Sprachmodell erzeugt darauf basierend eine Antwort oder ein neues Dokument.

```python
## WebBased Loader

from langchain_community.document_loaders import WebBaseLoader

## load all text
web_loader = WebBaseLoader(web_paths=("https://lilianweng.github.io/posts/2023-06-23-agent/",),)
web_loader.load()

```

```python
## WebBased Loader

from langchain_community.document_loaders import WebBaseLoader
import bs4
## load text only from aread where a css class "title" is
post_title_class = "post-title"
web_loader = WebBaseLoader(web_paths=("https://lilianweng.github.io/posts/2023-06-23-agent/",),
                           bs_kwargs=dict(parse_only=bs4.SoupStrainer(class_=(post_title_class))))
web_loader.load()
```
